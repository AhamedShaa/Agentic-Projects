```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as pdf:
        for page in pdf:
            text += page.get_text()
    return text

text_content = extract_text_from_pdf("Python_Datascience.pdf")
```

```python
document = {
    "title": "Python Data Science Book",
    "content": text_content,
    "type": "book",
    "source": "local"
}
```

```python
collection.insert_many(documents=[document])
```
